#### Joining sets

```python
# union

set1 = {"Nepal","India","China","Bhutan"}
set2 = {"India","Bangladesh"}

print(set1.union(set2))

#update adds element to the existing set
set1.update(set2)
set1
```

#### Intersection

```python
#intersection and intersection_update

setA = {"apple", "banana", "mango", "orange", "grape"}
setB = {"banana", "kiwi", "orange", "pineapple", "grape"}

setC = setA.intersection(setB)
print(setC)

setB.intersection_update(setA)
setB
```

#### Symmetric difference
- (AUB) - (AnB), all the elements except the common elements

```python
#symmetric difference i.e (AUB)-(AnB), all the elements except the common elements

setA = {"apple", "banana", "mango", "orange", "grape"}
setB = {"banana", "kiwi", "orange", "pineapple", "grape"}

sym = setA.symmetric_difference(setB)
sym1 = setA ^ setB
print(sym)
print(sym1)

#symmetric difference update
setA.symmetric_difference_update(setB)
print(setA)
```

#### isdisjoint()
- checks if two sets have items in common or not
- if no common items then the sets are said to be disjoint

```python
setA = {"apple", "banana", "mango", "orange", "grape"}
setB = {"banana", "kiwi", "orange", "pineapple", "grape"}

result = setA.isdisjoint(setB)

if result == True:
    print("The sets dont have items in common")
else:
    print("The sets have items in common")
```

#### Remove/Discard
- remove() raises error if element isnt present but discard() doesnt raise error

```python
setA = {"apple", "banana", "mango", "orange", "grape"}
setB = {"banana", "kiwi", "orange", "pineapple", "grape"}

#remove()
setA.remove("banana")
print(f"Set A after removing banana is {setA}")

#discard
setB.discard("kiwi")
print(f"Set B after removing banana is {setB}")
```

#### pop()
- pops a random value as sets are unordered

```python
setA = {"apple", "banana", "mango", "orange", "grape"}
popped_element = setA.pop()
print(f"The popped element is {popped_element}")
```

#### del 
- unlike other methods del is keyword that deletes the set entirely

```python
setA = {"apple", "banana", "mango", "orange", "grape"}
del setA

# print(setA) will throw error as the setA is deleted 
```

#### clear()
- clear the elements in the set

```python
setA = {"apple", "banana", "mango", "orange", "grape"}
setA.clear()
setA
```
