#### Creating dictionary

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(student)
```

####

### Accessing dictionary values

#### Accessing single values

student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(student["age"])
print(student.get("age"))

# if the key isnt present
print(student.get("age1")) ## return None if the key isnt present
print(student["age1"])  ## raises error if key isnt present




#### Accessing keys

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(student.keys())

```

#### Accessing items i.e key-value pair

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

for key,value in student.items():
    print(f"The value corresponding to the key {key} is {value}")
```

#### Changing values


```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(f"Before changing:\n{student}")
student["marks"] = 95

print(f"\nAfter changing: \n{student}")
```

#### Updating values
- update()
- update() changes the value if present else add new key-value pair to the dictionary
- can add multiple key value pairs

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

student.update({"marks":94}) #updates the existing value 
student.update({"college":"Orchid","semester":"7th"}) # adds new key-value pair

student
```

### Removing items

####  Using clear()

```python
student = {
 'name': 'Prajwal',
 'age': 21,
 'course': 'BSc CSIT',
 'marks': 94,
 'college': 'Orchid',
 'semester': '7th'
 }

student.clear()
student
```

#### Using popitem()
- removes the last inserted item

```python
student = {
 'name': 'Prajwal',
 'age': 21,
 'course': 'BSc CSIT',
 'marks': 94,
 'college': 'Orchid',
 'semester': '7th'
 }

student.popitem()
student
```

#### Using pop()
- removes the specified item


```python
student = {
 'name': 'Prajwal',
 'age': 21,
 'course': 'BSc CSIT',
 'marks': 94,
 'college': 'Orchid',
 'semester': '7th'
 }

student.pop("college")
student
```

#### Using del keyword
- removes item with specified key name
- can also delete the dictionary

```python
student = {
 'name': 'Prajwal',
 'age': 21,
 'course': 'BSc CSIT',
 'marks': 94,
 'college': 'Orchid',
 'semester': '7th'
 }

del student["semester"]

student

```

#### Copying a dictionary
- can be done by either copy() method or dict() function

```python
student = {
 'name': 'Prajwal',
 'age': 21,
 'course': 'BSc CSIT',
 'marks': 94,
 'college': 'Orchid',
 'semester': '7th'
 }


new_dict = student.copy()
print(f"Dictionary copied using copy() method : \n{new_dict}")

new_dict2 = dict(student)
print(f"\nDictionary copied using dict() function:\n{new_dict2}")
```

#### Nested Dictionary

```python
dict = {
    "book1":{"title":"Python",
             "published_year":"2025"},
    "book2":{"title":"C++",
             "published_year":"2023"},
    "book3":{"title":"Java",
             "published_year":"2024"}
}


##looping through nested dictionary

for key,value in dict.items():
    print(f"{key}:{value}")
    for x in value:
        print(value[x])


dict["book1"]["title"]
```
