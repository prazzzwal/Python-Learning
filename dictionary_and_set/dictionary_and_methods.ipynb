#### Creating dictionary

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(student)
```

### Accessing dictionary values

#### Accessing single values

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(student["age"])
print(student.get("age"))

# if the key isnt present
print(student.get("age1")) ## return None if the key isnt present
print(student["age1"])  ## raises error if key isnt present


```

#### Accessing keys

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(student.keys())

```

#### Accessing items i.e key-value pair

```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

for key,value in student.items():
    print(f"The value corresponding to the key {key} is {value}")
```

#### Changing values


```python
student = {
    "name": "Prajwal",
    "age": 21,
    "course": "BSc CSIT",
    "marks": 85
}

print(f"""Before changing:
      {student}""")
student["marks"] = 95

print(f"After changing: \n{student}")
```
